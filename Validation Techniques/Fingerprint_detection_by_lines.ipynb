### Fingerprint feature extraction - Lines

Idea: Utilising Hugh transform from OpenCV library to detect a fingerprint

Step 1: Image is converted to grayscale.
Step 2: Gaussian blur is applied to the grayscale image to reduce noise and enhance fingerprint features.
Step 3: Canny edge detection is employed to identify edges within the blurred image, potentially outlining ridges characteristic of fingerprints.
Step 4: Following edge detection, the Hough transform is utilized to detect lines corresponding to ridges in the image.
Step 5: If lines indicative of fingerprint ridges are found then this indicates a fingerprint.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_image(image, title="Image"):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def detect_fingerprint_by_lines(image_path):
    image = cv2.imread(image_path)
    display_image(image, "Original Image")
    # Step 1
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    display_image(gray, "Step 1: Grayscale Image")
    # Step 2
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    display_image(blurred, "Step 2: Blurred Image")
    # Step 3
    edges = cv2.Canny(blurred, 30, 150)
    display_image(edges, "Step 3: Edge Detection Result")
    # Step 4
    lines = cv2.HoughLines(edges, 1, np.pi / 180, 200)
    print(lines)
    # If lines are detected then it might be a fingerprint
    if lines is not None:
        print("Fingerprint detected!")
    else:
        print("No fingerprint detected.")
```

```python
# Method 1
detect_fingerprint_by_lines('../Datasets/Sokoto Coventry Fingerprint Dataset/Real/1__M_Left_index_finger.BMP')
```

Note: Line detection using Hough transform may not always work well for fingerprint images, especially if the ridges are not prominent or if the image quality is poor.
