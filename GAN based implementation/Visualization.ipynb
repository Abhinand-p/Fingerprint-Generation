```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from PIL import Image
```

```python
output_dir = "Generated_Images_NIST_Special_Dataset"
```

```python
image_files = sorted([f for f in os.listdir(output_dir) if f.endswith('.bmp')])
```

```python
def update(frame):
    im.set_array(frame)
    return im,
```

```python
# Creating animation
frames = []
for file in image_files:
    image_path = os.path.join(output_dir, file)
    image = Image.open(image_path)
    frames.append(np.array(image))

fig, ax = plt.subplots()
plt.axis("off")
im = ax.imshow(frames[0])
```

```python
ani = FuncAnimation(fig, update, frames=frames, blit=True)
ani.save('generated_images_animation_NIST_Special_Dataset.gif', writer='imagemagick', fps=10)
plt.show()
```
